# import libraries

```python
import cv2
from IPython import display
```

# read the image

```python
display.Image("new.jpg")
```

```python
image = cv2.imread("new.jpg")
cv2.waitKey(0)
```

# converting the original image to greyscale

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2.imshow("New", gray_image)
cv2.waitKey(0)
```

#  invert the new grayscale image

```python
inverted_image = 255 - gray_image
cv2.imshow("Inverted", inverted_image)
cv2.waitKey()
```

# blur the image by using the Gaussian Function in OpenCV

```python
blurred = cv2.GaussianBlur(inverted_image, (21, 21), 0)
```

# convert the image into a pencil sketch

```python
inverted_blurred = 255 - blurred
pencil_sketch = cv2.divide(gray_image, inverted_blurred, scale=256.0)
cv2.imshow("Sketch", pencil_sketch)
cv2.waitKey(0)
```
